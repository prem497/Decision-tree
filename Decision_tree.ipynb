```python
import pandas as pd # Library class/function for using table related datas
```

```python
dataset=pd.read_csv("50_Startups.csv") #upload and read csv file 
```

```python
dataset #view the data table
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.astype(int)
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent = dataset[["Profit"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion="absolute_error",splitter="best") # default parameter =squared_error so we can try other parameter to evaluate r_square value
regressor=regressor.fit(x_train,y_train) # regressor
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="DecisionTreeRegressionModel_final.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
